```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Restaurant_Reviews.tsv',sep = '\t')
df.head()
```

```python
df.notna().sum()
```

```python
df.info()
```

```python
x=df['Review']
```

```python
x
```

```python
y=df['Liked']
```

```python
y
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(x)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
tokenizer.document_count
```

```python
sequences=tokenizer.texts_to_sequences(x)
sequences
```

```python
from keras.utils import pad_sequences
from keras.layers import LSTM
from keras.layers import Embedding
```

```python
sent_length=50
```

```python
sequences=pad_sequences(sequences,padding='post',maxlen=sent_length)
```

```python
sequences
```

```python
len(sequences[999])
```

```python
from keras import Sequential
from keras.layers import Dense,SimpleRNN,Embedding,Flatten
from keras.layers import Dropout
```

```python
embedding__vector_features=40
voc_size=5000
```

```python
model=Sequential()
```

```python
from keras.layers.normalization.batch_normalization_v1 import BatchNormalization
model.add(Embedding(voc_size,embedding__vector_features,input_length=sent_length))
model.add(BatchNormalization())
model.add(Dropout(0.2))
```

```python
model.add(LSTM(100))
model.add(BatchNormalization())
model.add(Dropout(0.2))
```

```python
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
x_final=np.array(sequences)
y_final=np.array(y)
```

```python
x_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_final,y_final,test_size=0.3,random_state=22)
```

```python
model.fit(x_train,y_train,validation_data=(x_test,y_test),epochs=20,batch_size=32)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred[30]
```

```python
y_pred = model.predict(x_test)
y_pred = np.round(y_pred).astype(int)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
